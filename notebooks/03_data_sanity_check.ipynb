# Data Sanity Check

qwen2.5:7b-instruct 으로 가공한 TL_내과_통합_documents.json 데이터 점검

## 1) 데이터 null 값 확인

```python
from pathlib import Path
import json

# 1) Load TL_내과_통합_documents.json
json_path = Path("../data/processed/chroma_documents/TL_내과_통합_documents.json")
with json_path.open("r", encoding="utf-8-sig") as f:
    documents = json.load(f)

print(f"Loaded documents: {len(documents):,}")

# 2) Count unique ids
unique_id_count = len({doc.get("id") for doc in documents if doc.get("id") is not None})
print(f"Unique id count: {unique_id_count:,}")

# 3) Count empty list([]) in selected metadata fields
target_fields = ["질병명", "증상", "처치 방법", "환자 특성"]

empty_list_counts = {}
for field in target_fields:
    count = 0
    for doc in documents:
        metadata = doc.get("metadata", {}) or {}
        value = metadata.get(field)
        if isinstance(value, list) and len(value) == 0:
            count += 1
    empty_list_counts[field] = count

print("Empty list([]) counts in metadata fields:")
for field, count in empty_list_counts.items():
    print(f"- {field}: {count:,}")
```
